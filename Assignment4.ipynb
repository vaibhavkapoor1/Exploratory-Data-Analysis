```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib inline
#sns.set(color_codes=True)
```

```python
df=df.drop(['Longitude'],axis=1)
df.head(5)
```

```python
df.shape
```

```python
df=df.drop(['Latitude'],axis=1)
df.head(1)
```

```python
df=df.drop(['Locality Verbose'], axis=1)
```

```python
df.head(1)
```

```python
df=df.drop(['Locality'], axis=1)
```

```python
df.head()
```

```python
df=df.drop(['Rating color'], axis=1)
df.head(1)
```

```python
df=df.rename(columns={"Restaurant ID": "ID"})
df.head(1)
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

```python
ratings=final_df.groupby(['Aggregate rating','Votes']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Votes",data=ratings)
```
